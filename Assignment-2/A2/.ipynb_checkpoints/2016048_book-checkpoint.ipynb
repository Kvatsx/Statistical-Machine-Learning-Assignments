```python
import numpy as np
import matplotlib.pyplot as plt
from math
# import pandas as pd

def ReadData(filename):
    Arr = []
    Brr = []
    count = 0
    label = 1
    depth = 0
    with open(filename, 'r') as fp:
        line = fp.readlines()
        for row in line:
            # print(row)
            depth += 1
            row = list(map(float, row.split(" ")))
            Arr.append(row)
            # print(row)
            Brr.append(label)
            count += 1
            if count == 10:
                label += 1
                count = 0
    Arr = np.asarray(Arr)
    Brr = np.asarray(Brr)
    return Arr, Brr
```

```python
TrainX ,TrainY = ReadData('bookData.txt')
print(TrainX.shape)
print(TrainY.shape)
print(TrainY)

def Mean(trainx):
    arr = np.zeros((3, 3), dtype=float)
    for i in range(3):
        count = i*10
        sum = [0, 0, 0]
        for j in range(0, 10):
            sum[0] += trainx[i+j, 0]
            sum[1] += trainx[i+j, 1]
            sum[2] += trainx[i+j, 2]

        sum[0] /= 10
        sum[1] /= 10
        sum[2] /= 10
        arr[i, 0] = sum[0]
        arr[i, 1] = sum[1]
        arr[i, 2] = sum[2]
    # print(arr)
    return arr

def Variance(trainx, mean):
    arr = np.zeros((3, 3), dtype=float)
    for i in range(3):
        count = i*10
        sum = [0, 0, 0]
        for j in range(0, 10):
            sum[0] += ((trainx[i+j ,0]-mean[i, 0])**2)
            sum[1] += ((trainx[i+j, 2]-mean[i, 1])**2)
            sum[2] += ((trainx[i+j ,1]-mean[i, 2])**2)

        sum[0] /= 10
        sum[1] /= 10
        sum[2] /= 10
        arr[i, 0] = sum[0]
        arr[i, 1] = sum[1]
        arr[i, 2] = sum[2]
    # print(arr)
    return arr

def mean(trainx):
    sum = 0
    for i in range(10):
        sum += trainx[i, 0]
    print("M", sum/10)

# def CoVariance()

meanData = Mean(TrainX)
mean(TrainX)
varianceData = Variance(TrainX, meanData)
print(meanData)
print(varianceData)
```
